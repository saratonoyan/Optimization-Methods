```python
import numpy as np
import plotly.graph_objects as go
from scipy.optimize import minimize
```

```python
# ─── Objective function ───────────────────────────────────────────────────────

def f(x):                                                # define the objective function
    return (1/3) * x**4 + 2*x**2 - 6*x + 5             # f(x) = 1/3*x^4 + 2x^2 - 6x + 5
```

```python
# ─── Step 1: Set search parameters ───────────────────────────────────────────

a = 0                                                    # left bound of the interval
b = 3                                                    # right bound of the interval
e = 0.5                                                  # step size
```

```python
# ─── Step 2: Compute number of sample points ─────────────────────────────────

n = int((b - a) / e)                                     # compute number of intervals
print(f"Number of sample points: {n + 1}")               # print total number of points
```

```python
# ─── Step 3: Generate equally spaced x-values ────────────────────────────────

l_x = [a + i * e for i in range(n + 1)]                 # generate x_i = a + i*e
print("Sample x-values:", l_x)                           # print all sample x-values
```

```python
# ─── Step 4: Evaluate the function at each point ─────────────────────────────

l_y = [f(x) for x in l_x]                               # compute f(x) for every point
print("f(x) values:", [round(y, 4) for y in l_y])       # print rounded f(x) values
```

```python
# ─── Step 5: Find the approximate minimum ────────────────────────────────────

f_min = min(l_y)                                         # find the smallest f(x) value
x_min = l_x[l_y.index(f_min)]                           # find x that gives smallest f(x)
```

```python
# ─── Step 6: Find the exact minimum using scipy ──────────────────────────────

result      = minimize(f, x0=1.5, bounds=[(0, 3)])       # find exact minimum using scipy
min_x_exact = result.x[0]                                # extract exact x*
min_y_exact = result.fun                                  # extract exact f(x*)
```

```python
# ─── Step 7: Print comparison ────────────────────────────────────────────────

print("\n" + "=" * 55)                                   # print separator line
print("           RESULTS — METHOD OF ALTERNATIVES")      # print header
print("=" * 55)                                          # print separator line
print(f"  Approximate x*     : {x_min:.6f}")             # print approximate x
print(f"  Approximate f(x*)  : {f_min:.6f}")             # print approximate f(x)
print(f"  Exact x*           : {min_x_exact:.6f}")       # print exact x
print(f"  Exact f(x*)        : {min_y_exact:.6f}")       # print exact f(x)
print(f"  Error in x         : {abs(x_min - min_x_exact):.6f}")   # print x error
print(f"  Error in f(x)      : {abs(f_min - min_y_exact):.6f}")   # print f error
print("=" * 55)                                          # print separator line

# ─── Step 8: Prepare smooth curve ────────────────────────────────────────────

x_vals = np.linspace(0, 3, 400)                          # generate 400 points for smooth curve
y_vals = f(x_vals)                                        # evaluate function on smooth curve

# ─── Step 9: Create the figure ───────────────────────────────────────────────

fig = go.Figure()                                        # create new plotly figure

# ─── Step 10: Add shaded area under the curve ────────────────────────────────

fig.add_trace(go.Scatter(                                # add shaded fill under curve
    x=x_vals,                                            # x values
    y=y_vals,                                            # y values
    fill='tozeroy',                                      # fill down to y=0
    fillcolor='rgba(65, 105, 225, 0.07)',                # very light blue fill
    mode='none',                                         # no line for this trace
    showlegend=False,                                    # hide from legend
    hoverinfo='skip'                                     # skip hover for this trace
))

# ─── Step 11: Add the function curve ─────────────────────────────────────────

fig.add_trace(go.Scatter(                                # add function curve trace
    x=x_vals,                                            # x values for curve
    y=y_vals,                                            # y values for curve
    mode='lines',                                        # draw as line
    name='f(x) = ⅓x⁴ + 2x² − 6x + 5',                 # legend label
    line=dict(color='royalblue', width=3),               # thick royal blue line
    hovertemplate='x = %{x:.4f}<br>f(x) = %{y:.4f}<extra></extra>'  # hover text
))

# ─── Step 12: Add vertical dashed lines at each sample point ─────────────────

for xi, yi in zip(l_x, l_y):                            # loop over each sample point
    fig.add_shape(                                       # add vertical dashed line
        type='line',                                     # line shape
        x0=xi, y0=0,                                     # start at bottom
        x1=xi, y1=yi,                                    # end at f(x)
        line=dict(color='seagreen', width=1.5, dash='dot')  # green dotted line
    )

# ─── Step 13: Add the sampled points ─────────────────────────────────────────

fig.add_trace(go.Scatter(                                # add sampled points trace
    x=l_x,                                               # x values of sampled points
    y=l_y,                                               # y values of sampled points
    mode='markers',                                      # draw as markers only
    name=f'Sampled points (ε = {e})',                    # legend label
    marker=dict(                                         # marker style settings
        size=12,                                         # marker size
        color=l_y,                                       # color by f(x) value
        colorscale='Viridis',                            # color scale
        showscale=True,                                  # show color scale bar
        colorbar=dict(title='f(x)', x=1.02),             # color bar title
        line=dict(color='white', width=1.5)              # white border on markers
    ),
    hovertemplate='x = %{x:.4f}<br>f(x) = %{y:.6f}<extra></extra>'  # hover text
))

# ─── Step 14: Add the approximate minimum marker ─────────────────────────────

fig.add_trace(go.Scatter(                                # add approximate minimum
    x=[x_min],                                           # x position
    y=[f_min],                                           # y position
    mode='markers',                                      # draw as marker
    name=f'Approximate minimum<br>x* = {x_min:.4f},  f = {f_min:.6f}',  # legend
    marker=dict(size=18, color='orange', symbol='star',  # large orange star
                line=dict(color='darkorange', width=2)), # dark orange border
    hovertemplate=f'Approximate<br>x* = {x_min:.4f}<br>f(x*) = {f_min:.6f}<extra></extra>'
))

# ─── Step 15: Add the exact minimum marker ───────────────────────────────────

fig.add_trace(go.Scatter(                                # add exact minimum trace
    x=[min_x_exact],                                     # x position
    y=[min_y_exact],                                     # y position
    mode='markers',                                      # draw as marker
    name=f'Exact minimum<br>x* = {min_x_exact:.4f},  f = {min_y_exact:.6f}',  # legend
    marker=dict(size=18, color='crimson', symbol='star', # large red star
                line=dict(color='darkred', width=2)),    # dark red border
    hovertemplate=f'Exact<br>x* = {min_x_exact:.4f}<br>f(x*) = {min_y_exact:.6f}<extra></extra>'
))

# ─── Step 16: Add result annotation box ──────────────────────────────────────

fig.add_annotation(                                      # add annotation box
    x=x_min, y=f_min,                                    # position at approximate minimum
    text=(f'<b>Approximate</b><br>'
          f'x* = {x_min:.4f}<br>'
          f'f(x*) = {f_min:.6f}<br>'
          f'<b>Exact</b><br>'
          f'x* = {min_x_exact:.4f}<br>'
          f'f(x*) = {min_y_exact:.6f}<br>'
          f'<b>Error in x</b> = {abs(x_min - min_x_exact):.6f}<br>'
          f'<b>Error in f</b> = {abs(f_min - min_y_exact):.6f}'),  # full comparison text
    showarrow=True,                                      # show arrow
    arrowhead=2,                                         # arrow style
    arrowcolor='orange',                                 # arrow color
    ax=120, ay=-120,                                     # arrow offset
    bgcolor='white',                                     # white background
    bordercolor='orange',                                # orange border
    borderwidth=2,                                       # border width
    font=dict(size=11, color='#333')                     # font style
)

# ─── Step 17: Apply layout settings ──────────────────────────────────────────

fig.update_layout(                                       # apply layout settings
    title=dict(                                          # title settings
        text=(f'<b>Method of Alternatives</b> — Uniform Scan on [0, 3]<br>'
              f'<sup>Approximate: x* = {x_min:.4f}, f = {f_min:.6f}  |  '
              f'Exact: x* = {min_x_exact:.4f}, f = {min_y_exact:.6f}  |  '
              f'Error in x = {abs(x_min - min_x_exact):.6f}</sup>'),
        font=dict(size=16),                              # title font size
        x=0.5,                                           # center the title
        xanchor='center'                                 # anchor to center
    ),
    xaxis=dict(title='x', showgrid=True,                 # x-axis settings
               gridcolor='rgba(200,200,200,0.3)',
               zeroline=True, zerolinecolor='rgba(0,0,0,0.2)'),
    yaxis=dict(title='f(x)', showgrid=True,              # y-axis settings
               gridcolor='rgba(200,200,200,0.3)'),
    hovermode='closest',                                 # hover mode
    template='plotly_white',                             # white theme
    legend=dict(yanchor='top', y=0.99,                   # legend position
                xanchor='left', x=0.01,
                bgcolor='rgba(255,255,255,0.9)',
                bordercolor='lightgray', borderwidth=1),
    width=950, height=580,                               # figure dimensions
    plot_bgcolor='white', paper_bgcolor='white'          # background colors
)

fig.show()                                               # display interactive plot

import pandas as pd                         # import pandas for table display
from IPython.display import display         # import display for rich output


# ─── Iteration Table — Method of Alternatives ─────────────────────────────────

print("\n" + "=" * 60)                      # print separator line
print("          ITERATION TABLE — METHOD OF ALTERNATIVES")  # print header
print("=" * 60)                             # print separator line

alt_table = []                              # list to store table rows

for i, (xi, yi) in enumerate(zip(l_x, l_y)):  # loop over each sample point
    alt_table.append({                      # add row to table
        'Iteration': i + 1,                 # iteration number
        'x (argument)': round(xi, 6),       # x value
        'f(x) (result)': round(yi, 6),      # f(x) value
        'Is Minimum': '✅ YES' if xi == x_min else ''  # mark the minimum
    })

alt_df = pd.DataFrame(alt_table)           # convert to dataframe
display(alt_df)                             # display the table

print(f"\n  Total iterations  : {n + 1}")   # print total sample points
print(f"  Best argument x*  : {x_min:.6f}") # print best x
print(f"  Best result f(x*) : {f_min:.6f}") # print best f(x)
print(f"  Exact x*          : {min_x_exact:.6f}")  # print exact x
print(f"  Exact f(x*)       : {min_y_exact:.6f}")  # print exact f(x)
print(f"  Error in x        : {abs(x_min - min_x_exact):.6f}")  # print x error
print(f"  Error in f(x)     : {abs(f_min - min_y_exact):.6f}")  # print f error
print("=" * 60)                             # print separator line
```
