```python
import numpy as np
import plotly.graph_objects as go
from scipy.optimize import minimize
```

```python
# ─── Objective function ───────────────────────────────────────────────────────

def f(x):                                                # define the objective function
    return (1/3) * x**4 + 2*x**2 - 6*x + 5             # f(x) = 1/3*x^4 + 2x^2 - 6x + 5
```

```python
# ─── Step 1: Set search parameters ───────────────────────────────────────────

a   = 0                                                  # left bound of the interval
b   = 3                                                  # right bound of the interval
eps = 0.5                                                # convergence tolerance
d   = 0.125                                              # probe offset
```

```python
# ─── Step 2: Initialize storage ──────────────────────────────────────────────

iteration_midpoints = []                                 # list to store midpoints
intervals           = []                                 # list to store interval bounds
all_x1              = []                                 # list to store all x1 probes
all_x2              = []                                 # list to store all x2 probes
iteration           = 0                                  # iteration counter
```

```python
# ─── Step 3: Run the bisection loop ──────────────────────────────────────────

while abs(b - a) > eps:                                  # loop until interval small enough
    intervals.append((a, b))                             # record current interval
    midpoint = (a + b) / 2                               # compute midpoint
    iteration_midpoints.append(midpoint)                 # record midpoint

    x1 = (a + b - d) / 2                                 # compute left probe point
    x2 = (a + b + d) / 2                                 # compute right probe point
    all_x1.append(x1)                                    # store left probe
    all_x2.append(x2)                                    # store right probe

    if f(x1) < f(x2):                                    # compare function values
        b = x2                                           # shrink interval from right
    else:                                                # otherwise
        a = x1                                           # shrink interval from left

    iteration += 1                                       # increment counter
```

```python
# ─── Step 4: Compute approximate minimum ─────────────────────────────────────

xmin  = (a + b) / 2                                      # final approximate minimum
f_min = f(xmin)                                          # final approximate f(x*)
```

```python
# ─── Step 5: Find the exact minimum using scipy ──────────────────────────────

result      = minimize(f, x0=1.5, bounds=[(0, 3)])       # find exact minimum using scipy
min_x_exact = result.x[0]                                # extract exact x*
min_y_exact = result.fun                                  # extract exact f(x*)
```

```python
# ─── Step 6: Print comparison ────────────────────────────────────────────────

print("\n" + "=" * 55)                                   # print separator line
print("              RESULTS — BISECTION METHOD")         # print header
print("=" * 55)                                          # print separator line
print(f"  Approximate x*     : {xmin:.6f}")              # print approximate x
print(f"  Approximate f(x*)  : {f_min:.6f}")             # print approximate f(x)
print(f"  Exact x*           : {min_x_exact:.6f}")       # print exact x
print(f"  Exact f(x*)        : {min_y_exact:.6f}")       # print exact f(x)
print(f"  Error in x         : {abs(xmin - min_x_exact):.6f}")    # print x error
print(f"  Error in f(x)      : {abs(f_min - min_y_exact):.6f}")   # print f error
print(f"  Iterations         : {iteration}")             # print iterations used
print(f"  Function evals     : {2 * iteration}")         # print function evaluations
print("=" * 55)                                          # print separator line
```
